```python
marks = {'Andy':88,'Amy':66,'James':90,'Jules':55,'Arthur':77}
chartnames = marks.keys()
chartvalues = marks.values()

#Bar Chart
plt.bar(chartnames,chartvalues)

plt.title('Student Grade Bar Chart')
plt.xlabel('Student Names')
plt.ylabel('Student Grades')

plt.show()

#Pie Chart
plt.pie(chartvalues, labels = chartnames)

plt.title('Student Grades Pie Chart')
plt.legend(loc = 'upper left', bbox_to_anchor=(1.05,1.0), title = 'Legend')

plt.show()
```

```python
Q6x = [1,3,4,5,7]
Q6y = [1,3,6,7,10]
Q6y2 = [1,0,4,2,3]
histValues = [1,2,4,4,4,3,2,1,2,3,5,6,7,8,8,7,7,7,5,4,3,3]

#Plot 1 - "One Subplot has multiple lines"
plt.subplot(2,3,1)
plt.plot(Q6x,Q6y)
plt.plot(Q6x,Q6y2)
plt.title("Multiple Lines")

#Plot 2 - "One subplot is a bar chart"
plt.subplot(2,3,2)
plt.bar(Q6x,Q6y)
plt.title("Bar Graph")

#Plot 3 - "One subplot is a pie chart"
plt.subplot(2,3,3)
plt.pie(Q6y, labels = Q6x)
plt.title("Pie Chart")

#Plot 4 - "One subplot has a grid, you define it yourself"
plt.subplot(2,3,4)
plt.plot(Q6y,Q6x) #switching x and y for variation's sake
plt.grid(color = "red", linewidth = 3, linestyle = '--')
plt.title("Now with grids")

#PLot 5 - "One subplot is a scatter plot"
plt.subplot(2,3,5)
plt.scatter(Q6x,Q6y)
plt.title("Scatterplot")

#Plot 6 - "One subplot is a histogram plot"
plt.subplot(2,3,6)
plt.hist(histValues)
plt.title("Histogram")

plt.show()
```
